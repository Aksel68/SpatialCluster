```python
from PyNetworkD3 import ForceGraph, ArcDiagram, AdjacencyMatrix, RadialDiagram
```

# Testing with random generated dataset

```python
from networkx.generators.random_graphs import erdos_renyi_graph
from random import randint

n = 100
p = 0.01
g = erdos_renyi_graph(n, p)

dataset = {"nodes": [], "links": []}
for node in g.nodes:
    dataset["nodes"].append({"id": node})
for edge in g.edges:
    dataset["links"].append({"source": edge[0], "target": edge[1], "value": randint(1, 100)})
```

```python
graph = ForceGraph(dataset, width=900, height=500, radio=4, force_link=10, force_collision=10,
                   force_simulation = -100, tooltip=["id"], view_box=True,
                   canvas=False, bounding_box=True)
graph.export("source.html")
graph
```

```python
graph = ForceGraph(dataset, width=1800, height=500, radio=4, force_link=10, force_collision=10,
                   force_simulation = -100,
                   tooltip=["id"], view_box=True, canvas=True, bounding_box=True)
graph.export("source.html")
graph
```

```python
matrix = AdjacencyMatrix(dataset, size=500, view_box=False, bidirrectional=False)
matrix
```

```python
matrix = AdjacencyMatrix(dataset, size=500, view_box=False, bidirrectional=True)
matrix
```

```python
arc = ArcDiagram(dataset, width=800, height=300, radio=5)
arc
```

```python
radial = RadialDiagram(dataset, size=500, view_box=False)
radial
```

# Testing with Data Research Network

```python
from urllib.request import urlopen, Request

url = "https://raw.githubusercontent.com/holtzy/D3-graph-gallery/master/DATA/data_researcherNetwork.json"

httprequest = Request(url, headers={"Accept": "application/json"})

with urlopen(httprequest) as response:
    dataset2 = response.read().decode()
```

```python
graph3 = ForceGraph(dataset2, width=900, height=400, tooltip=["id, n"],
                    force_collision=20, force_simulation=-5, force_link=10, canvas=True)
graph3.radio = 2
graph3
```

```python
arc = ArcDiagram(dataset2, width=500, radio=3)
arc
```

```python
matrix = AdjacencyMatrix(dataset2, size=800, view_box=False, bidirrectional=True)
matrix
```
